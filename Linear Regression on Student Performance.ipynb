### Simple Linear Regression - Study Hours vs Student Score - Student_Performance.csv dataset

### dataset par aap study hours ke basis par student ke scores predict karein ge using Simple Linear Regression. Complete ML workflow follow karein including preprocessing, training aur evaluation


```python
import pandas as pd
data = pd.read_csv('Student_Performance.csv')
```

```python
data.duplicated().sum()
```

```python
data.head()
```

```python
data.drop_duplicates(inplace=True)
```

```python
data.isna().sum()
```

```python
x = data[['Hours Studied']]
y = data['Performance Index']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
model = LinearRegression()
```

```python
model.fit(x_train, y_train)
```

```python
y_pred_test = model.predict(x_test)
```

```python
r2_score(y_test, y_pred_test)
```

```python
def score_predictor(score):
    return model.predict([[score]])
```

```python
a = int(input("Enter your studied hours: "))
print(f'Your Predicted Score is: {score_predictor(a)[0]:.2f}')
```
